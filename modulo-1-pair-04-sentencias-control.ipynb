EJERCICIOS: if - else 

```python
#Ejercicio 1.

num = int(input("Pon un numero"))

if num > 0:
    print("Es positivo")
elif num < 0:
    print("Es negativo")
else: 
    print("Es 0")       
```

```python
#Ejercicio 2.

num2 = int(input("Pon un numero"))

if num2 % 2 == 0:
    print("Es par")
else: 
    print("Es impar")    
```

```python
#Ejercicio 3.

examen = int(input("Que nota has tenido?"))

if examen >= 60:
    print("Aprobado")
else:
    print("Suspendido")    
```

```python
#Ejercicio 4.

x = int(input("Primer numero"))
y = int(input("Segundo numero"))

if x > y:
    print("Es mayor", x)

else:
    print("Es mayor", y)
```

```python
#Ejercicio 5.

letra = ["a", "e", "i", "o", "u"]
letra2 = input("Pon una letra")

if letra2 in letra:
    print("Es vocal")
    
else:
    print("Es consonante")
```

```python
#Ejercicio 6.

compra = int(input("¿Cuanto te ha costado?"))

if compra >= 100:
    print("Descuento del 10%:", compra * 0.10, "€")
else:
    print("No hay descuento")  
```

EJERCICIOS: While

```python
#Ejercicio 1.

contador = 1
suma = 0

while contador <= 10:
    suma += contador
    contador += 1
    
print("La suma de los 10 primeros numeros es:", suma)
```
