Ejercicio 3 Determinar si una cadena es un palíndromo.

Descripción: Crea una función llamada es_palindromo que tome una cadena como parámetro y devuelva True si es un palíndromo, o False si no lo es.

```python
def es_palindromo (texto):
    texto_invertido = texto[::-1]
    if texto == texto_invertido:
        return True

    else:
        return False
    
es_palindromo("rana")
```

Ejercicio 4 Calcular el promedio de una lista de números.

Descripción: Crea una función llamada calcular_promedio que tome una lista de números como parámetro y devuelva el promedio de todos sus elementos.

```python
lista_numeros = [1, 2, 3, 4, 5, 7, 10, 55]
def calcular_promedio(lista_numero):
    promedio = sum(lista_numero) / len(lista_numero)
    return promedio

calcular_promedio(lista_numeros)
```

Ejercicio 5 Calcular el interés compuesto.

Descripción: Crea una función llamada calcular_interes_compuesto que tome el capital inicial, la tasa de interés y el número de períodos como parámetros y devuelva el beneficio final después del interés compuesto.
A = P(1 + r)^n
A: Capital final (monto total).
P: Capital inicial (principal).
r: Tasa de interés por período (en decimal).
n: Número de períodos.

```python
def calcular_interes_compuesto (capital, tasa, num_periodo):

    capital_final = capital * (1 + tasa)** num_periodo
    return capital_final

calcular_interes_compuesto(5, 50, 2)

    
```
