calcular la suma de los primeros 10 numeros utilizando un bucle for:
Descripción: Utiliza un bucle for para calcular la suma de los primeros 10 números naturales (1 + 2 + 3 + ... + 10).


```python
suma=0
for n in  range(1,11):
   suma=suma+n
print(suma)
```

Calcular el producto de los elementos de una lista
Descripción: Dada una lista de números, utiliza un bucle for para calcular el producto de todos los elementos.

```python
numeros=[1,2,3,4,9,8,7,5,3]
num=1
for n in numeros:
    num=num*n
print(num)    
```


Imprimir los números primos del 1 al 50 utilizando un bucle for

Descripción: Utiliza un bucle for para imprimir los números primos del 1 al 50 en orden ascendente.

```python

for prim in range(2,51):
    es_primo=True
    for i in range(2,prim):
        if prim &i ==0:
         es_primo=False
         break
    if es_primo:
        print(prim)
    
```
