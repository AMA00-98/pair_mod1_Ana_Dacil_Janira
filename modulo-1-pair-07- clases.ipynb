```python
class Empleados:
    def __init__(self,nombre,apellidos,edad,posicion,año,vacaciones,herramientas = ["Python", "Excel"]):
        self.nombre=nombre
        self.apellidos=apellidos
        self.edad=edad
        self.posicion=posicion
        self.año=año
        self.vacaciones=vacaciones
        self.herramientas=herramientas

        
    def descripcion(self):
    
        return(f"el empleado {self.nombre} {self.apellidos}, puesto {self.posicion}, en el año {self.año} le quedan {self.vacaciones} dias de vacaciones")
    
    def cambiando_posicion(self):
        if "Python" in self.herramientas:
            return "Ya sabes Python, estudia Excel"
        
        else:
            return "Estudia Python"
        
    
```

```python
empleado1=Empleados("paco","garcia",34,"peon",2026,18,["python"])
```

```python
print(empleado1.descripcion())
```

```python
cambiando_posicion("Python")
```
